# Functions that return `None`

```python
# We're used to nesting functions:

numbers = []
line = '1,2,3,4,5'
n = line.split(',')
numbers.append(int(n[0])) # list indexing nest inside int conversion nested inside list append
numbers
```

```python
# Same principle, except in this case we can string functions together

file = open('transcript_data.csv')
header = file.readline().strip().split(',')
file.close()
print(header)
```

```python
# But this doesn't work with append(), because
# it modifies the list in place, and returns the value none

old_list = [0, 1, 2, 3, 4]

new_list = old_list.append(5) # returned value of append is None

print(new_list)
```

```python
# Functions can return None or lack return statements altogether (implied return None)

def some_function(an_argument):
    # modifies a list
    return None
```

# Plots

`Matplotlib` is the major plotting package in Python

1) Learning plots is not like learning Python basics â€“ you learn "recipes" to make plots, rather than reasoning from first principles. Like learning Illustrator.

2) Learn how to make quick simple plots, look up examples of more complex plots when you need to.

```python
# Setting up plotting in a notebook
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# A simple example - easy to make simple plots

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('some numbers')
plt.axis([0, 6, 0, 20])
```

```python
# Easy to plot more than one line 

x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
a = [2, 4, 6, 8, 10, 12]
b = [1, 2, 3, 4, 5, 6]
plt.plot(x, y, '-o', a, b, '-^')
plt.ylabel('X values')
plt.xlabel('Y values')
```

```python
# A colorful example from the matplotlib tutorials
import numpy as np
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data = data)
```

```python
# Make your own xkcd plots

with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.bar([0, 1], [0, 100], 0.25)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 1])
    ax.set_xlim([-0.5, 1.5])
    ax.set_ylim([0, 110])
    ax.set_xticklabels(['CONFIRMED BY\nEXPERIMENT', 'REFUTED BY\nEXPERIMENT'])
    plt.yticks([])

    plt.title("CLAIMS OF SUPERNATURAL POWERS")

    fig.text(
        0.5, -0.05,
        '"The Data So Far" from xkcd by Randall Monroe',
        ha='center')

```
